# Forward Propagation using Dot Product

```python
input_data = np.array([2,3])
weights = {'node_0':np.array([1,1]),
           'node_1':np.array([-1,1]),
           'output':np.array([2,-1])}
node_0_value = (input_data*weights['node_0']).sum()
node_1_value = (input_data*weights['node_1']).sum()
```

```python
hidden_layer_values = np.array([node_0_value, node_1_value])
print(hidden_layer_values)
```

```python
output = (hidden_layer_values * weights['output']).sum()
print(output)
```

## Try with a different model structure

```python
# Import necessary library
import numpy as np
```

```python
# Define the input data as a NumPy array with two values
input_data = np.array([2,3])

# Define the weights for each node in the neural network
weights = {'node_0':np.array([1,1]),
           'node_1':np.array([-1,1]),
           'node_2':np.array([1,-1]), # this node is added to create a 2 by 3 model structure
           'output':np.array([2,-1,1])}

# Calculate the value of node_0 in the hidden layer
node_0_value = (input_data*weights['node_0']).sum()

# Calculate the value of node_1 in the hidden layer
node_1_value = (input_data*weights['node_1']).sum()

# Calculate the value of node_2 in the hidden layer
node_2_value = (input_data*weights['node_2']).sum()
```

```python
# Combine the values of the three hidden layer nodes into an array
hidden_layer_values = np.array([node_0_value, node_1_value, node_2_value])

# Print the values of the hidden layer nodes
print(hidden_layer_values)
```

```python
# Calculate the output of the neural network by multiplying hidden_layer_values with output weights
output = (hidden_layer_values * weights['output']).sum()

# Print the final output of the neural network
print(output)
```
